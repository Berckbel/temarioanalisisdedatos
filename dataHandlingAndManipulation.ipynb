<h3>dataframe handling and manipulation</h3>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('data/datasets/weatherAUS.csv')
```

```python
dataframe.head(5)
```

```python
print("Hay {0[0]} filas y {0[1]} columnas. \n".format(dataframe.shape))
```

```python
dataframe.describe()
```

```python
#----datos float------
mintemp_mean = dataframe['MinTemp'].mean()
maxtemp_mean = dataframe['MaxTemp'].mean()
rainfall_mean = dataframe['Rainfall'].mean()
evaporation_mean = dataframe['Evaporation'].mean()
sunshine_mean = dataframe['Sunshine'].mean()
windgustspeed_mean = dataframe['WindGustSpeed'].mean()
windspeed9am_mean = dataframe['WindSpeed9am'].mean()
windspeed3pm_mean = dataframe['WindSpeed3pm'].mean()
humiditty9am_mean = dataframe['Humidity9am'].mean()
humidity3pm_mean = dataframe['Humidity3pm'].mean()
pressure9am_mean = dataframe['Pressure9am'].mean()
pressure3pm_mean = dataframe['Pressure3pm'].mean()
cloud9am_mean = dataframe['Cloud9am'].mean()
cloud3pm_mean = dataframe['Cloud3pm'].mean()
temp9am_mean = dataframe['Temp9am'].mean()
temp3pm_mean = dataframe['Temp3pm'].mean()

#---object(string)(direccion)
windgustdir_mode = dataframe['WindGustDir'].mode()[0]
winddir9am_mode = dataframe['WindDir9am'].mode()[0]
winddir3pm_mode = dataframe['WindDir3pm'].mode()[0]

#---objetc(string)(boolean) // mode = No
raintoday_mode = dataframe['RainToday'].mode()[0]
raintomorrow_mode = dataframe['RainTomorrow'].mode()[0]
```

```python
#----datos float------
dataframe['MinTemp'] = dataframe['MinTemp'].replace(np.nan, mintemp_mean)
dataframe['MaxTemp'] = dataframe['MaxTemp'].replace(np.nan, maxtemp_mean)
dataframe['Rainfall'] = dataframe['Rainfall'].replace(np.nan, rainfall_mean)
dataframe['Evaporation'] = dataframe['Evaporation'].replace(np.nan, evaporation_mean)
dataframe['Sunshine'] = dataframe['Sunshine'].replace(np.nan, sunshine_mean)
dataframe['WindGustSpeed'] = dataframe['WindGustSpeed'].replace(np.nan, windgustspeed_mean)
dataframe['WindSpeed9am'] = dataframe['WindSpeed9am'].replace(np.nan, windspeed9am_mean)
dataframe['WindSpeed3pm'] = dataframe['WindSpeed3pm'].replace(np.nan, windspeed3pm_mean)
dataframe['Humidity9am'] = dataframe['Humidity9am'].replace(np.nan, humiditty9am_mean)
dataframe['Humidity3pm'] = dataframe['Humidity3pm'].replace(np.nan, humidity3pm_mean)
dataframe['Pressure9am'] = dataframe['Pressure9am'].replace(np.nan, pressure9am_mean)
dataframe['Pressure3pm'] = dataframe['Pressure3pm'].replace(np.nan, pressure3pm_mean)
dataframe['Cloud9am'] = dataframe['Cloud9am'].replace(np.nan, cloud9am_mean)
dataframe['Cloud3pm'] = dataframe['Cloud3pm'].replace(np.nan, cloud3pm_mean)
dataframe['Temp9am'] = dataframe['Temp9am'].replace(np.nan, temp9am_mean)
dataframe['Temp3pm'] = dataframe['Temp3pm'].replace(np.nan, temp3pm_mean)

#---object(string)(direccion)
dataframe['WindGustDir'] = dataframe['WindGustDir'].replace(np.nan, windgustdir_mode)
dataframe['WindDir9am'] = dataframe['WindDir9am'].replace(np.nan, winddir9am_mode)
dataframe['WindDir3pm'] = dataframe['WindDir3pm'].replace(np.nan, winddir3pm_mode)

#---objetc(string)(boolean)
dataframe['RainToday'] = dataframe['RainToday'].replace(np.nan, raintoday_mode)
dataframe['RainTomorrow'] = dataframe['RainTomorrow'].replace(np.nan, raintomorrow_mode)
```
