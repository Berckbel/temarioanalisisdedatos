<h3>Web Scraping</h3>

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://es.wikipedia.org/wiki/Alan_Turing')
res.status_code
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
main_title = soup.find("span", class_="mw-page-title-main")
main_title.get_text()
```

<h3>API Recolection</h3>

```python
import requests
import json
```

```python
res = requests.get('https://pokeapi.co/api/v2/pokemon/pikachu')
res.status_code
```

```python
with open('data/recolection/pikachu.json', mode='w') as pikachu_obj:
    json_obj = json.loads(res.content)
    json.dump(json_obj, pikachu_obj, ensure_ascii=False, indent=4)
```
